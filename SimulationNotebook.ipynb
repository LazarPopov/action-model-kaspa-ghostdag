```python
import random
import networkx as nx
import matplotlib.pyplot as plt

###############################################################################
#                     Domain of Possible Worlds Setup                         #
###############################################################################

def all_unique_worlds():
    """
    Returns the list of the 8 unique worlds for the three propositions.
    Each world is represented as a dict with keys:
         'contract', 'a_contract', 'b_contract'
    """
    worlds = []
    for contract in [False, True]:
        for a_contract in [False, True]:
            for b_contract in [False, True]:
                worlds.append({
                    'contract': contract,
                    'a_contract': a_contract,
                    'b_contract': b_contract
                })
    return worlds

def get_domain():
    """
    Returns a dictionary mapping world IDs (0 to 7) to their truth assignments.
    """
    unique = all_unique_worlds()
    domain = {i: unique[i] for i in range(len(unique))}
    return domain

###############################################################################
#                              Agent Definition                               #
###############################################################################

class Agent:
    def __init__(self, agent_id, domain):
        self.agent_id = agent_id
        self.domain = domain  # the fixed domain of worlds
        # Initially, the agent considers every world in the domain possible.
        self.possible_worlds = set(domain.keys())
        # Keep track of which propositions this agent has already broadcast.
        self.broadcasted = set()
    
    def update_epistemic_state(self, announcement):
        """
        Given an announcement (e.g. {"contract": True}), update the epistemic state
        by keeping only those worlds that satisfy the announcement.
        """
        self.possible_worlds = {w for w in self.possible_worlds 
                                if all(self.domain[w][prop] == val for prop, val in announcement.items())}
    
    def __str__(self):
        return f"Agent {self.agent_id} (Possible worlds: {sorted(self.possible_worlds)})"

###############################################################################
#                     Agent Network & Visualization Helpers                     #
###############################################################################

def generate_agents(n, domain):
    """Generate n Agent objects with IDs 1..n, all using the same domain."""
    return [Agent(agent_id=i, domain=domain) for i in range(1, n+1)]

def generate_random_connections(n, p=0.3):
    """
    Generates an undirected, random connection network between n agents.
    Each pair of agents (i, j) is connected with probability p.
    
    Returns a dictionary of adjacency lists, e.g.:
         {1: [2, 4], 2: [1, 3], ...}
    """
    connections = {i: [] for i in range(1, n+1)}
    for i in range(1, n):
        for j in range(i+1, n+1):
            if random.random() < p:
                connections[i].append(j)
                connections[j].append(i)
    return connections

def visualize_agent_network(agent_ids, connections):
    """
    Visualizes the static agent connection network.
    """
    G = nx.Graph()
    for a in agent_ids:
        G.add_node(a)
    for agent, neighbors in connections.items():
        for nb in neighbors:
            G.add_edge(agent, nb)
    pos = nx.spring_layout(G, seed=42)
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color="lightblue",
            node_size=1400, font_size=10, font_weight="bold",
            edge_color="gray", width=2.0)
    plt.title("Agent Connection Network")
    plt.axis("off")
    plt.show()

def visualize_combined_kripke(agents, domain, tick):
    """
    Visualizes the combined Kripke model.
    
    - The nodes represent the 8 unique worlds (from the domain).  
    - For each agent, we draw a directed edge (arrow) between every pair of worlds 
      in that agent's current possible worlds.
      Each agents arrows are drawn in a distinct color.
    """
    G = nx.DiGraph()  # Use a directed graph so arrows can be drawn.
    
    # Add nodes for each world in the domain.
    for wid, assignment in domain.items():
        label = f"W{wid}\nC:{'T' if assignment['contract'] else 'F'}\n" \
                f"a:{'T' if assignment['a_contract'] else 'F'}\n" \
                f"b:{'T' if assignment['b_contract'] else 'F'}"
        G.add_node(wid, label=label)
    
    # Define colors for agents.
    color_list = ["red", "green", "blue", "orange", "violet", "brown"]
    agent_colors = {}
    for agent in agents:
        agent_colors[agent.agent_id] = color_list[(agent.agent_id - 1) % len(color_list)]
    
    # For each agent, add directed edges for its accessibility relation.
    # (Since the relation is complete over the set of possible worlds the agent considers possible,
    # add an edge from every world to every other world in that set.)
    agent_edges = {agent.agent_id: [] for agent in agents}
    for agent in agents:
        pw = sorted(list(agent.possible_worlds))
        for i in range(len(pw)):
            for j in range(len(pw)):
                if pw[i] != pw[j]:
                    agent_edges[agent.agent_id].append((pw[i], pw[j]))
    
    pos = nx.spring_layout(G, seed=42)
    node_labels = nx.get_node_attributes(G, "label")
    
    plt.figure(figsize=(10, 8))
    nx.draw_networkx_nodes(G, pos, node_color="lightgray", node_size=1500)
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8)
    
    # Draw each agent's accessibility edges in its color.
    for agent in agents:
        edges = agent_edges[agent.agent_id]
        nx.draw_networkx_edges(G, pos, edgelist=edges,
                               edge_color=agent_colors[agent.agent_id],
                               arrows=True, arrowstyle="->", arrowsize=10,
                               width=2, alpha=0.7, label=f"Agent {agent.agent_id}")
    
    plt.title(f"Combined Kripke Model at Tick {tick}")
    plt.axis("off")
    # Create a legend for agent colors.
    for agent in agents:
        plt.plot([], [], color=agent_colors[agent.agent_id], label=f"Agent {agent.agent_id}")
    plt.legend(loc="best")
    plt.show()

###############################################################################
#                           Epistemic Simulation                              #
###############################################################################

def epistemic_simulation(num_agents=5, num_ticks=5, p_connection=0.3):
    """
    Runs a simulation where:
      - The domain is fixed to the 8 unique worlds.
      - Each agent initially considers all worlds possible.
      - An agent network is generated and visualized.
      - Predefined epistemic actions occur at designated ticks:
            * The contract owner learns and broadcasts that {'contract': True}.
            * Two different agents broadcast that {'a_contract': True} and {'b_contract': True}.
      - Additionally, at every tick, if an agent has "learned" a proposition
        (i.e. its epistemic state is now such that all its possible worlds satisfy that proposition)
        and it has not yet broadcast that fact, then it schedules a broadcast for the next tick.
      - When a broadcast is processed, each neighbor updates its epistemic state accordingly.
      - Finally, after all updates for the tick, the combined Kripke model is visualized.
    """
    domain = get_domain()
    agents = generate_agents(num_agents, domain)
    connections = generate_random_connections(num_agents, p=p_connection)
    agent_ids = [a.agent_id for a in agents]
    
    # Visualize the static agent connection network.
    visualize_agent_network(agent_ids, connections)
    
    # --- Randomly assign roles ---
    # Randomly choose a contract owner and two signers; all roles must be distinct.
    contract_owner = random.choice(agent_ids)
    remaining = list(set(agent_ids) - {contract_owner})
    signer1, signer2 = random.sample(remaining, 2)
    print(f"Contract owner is Agent {contract_owner}")
    print(f"Signer 1 is Agent {signer1}")
    print(f"Signer 2 is Agent {signer2}")
    
    # Predefined epistemic actions.
    actions = {
        1: (contract_owner, {"contract": True}),
        2: (contract_owner, {"contract": True}),
        3: (signer1, {"a_contract": True}),
        4: (signer2, {"b_contract": True})
    }
    
    # Dictionary to schedule dynamic broadcast actions.
    # Key: tick number; Value: list of (agent_id, announcement) tuples.
    scheduled_broadcasts = {}
    
    for tick in range(1, num_ticks + 1):
        print(f"\n=== TICK {tick} ===")
        
        # Process any scheduled broadcast actions for this tick.
        if tick in scheduled_broadcasts:
            for agent_id, announcement in scheduled_broadcasts[tick]:
                source = next((a for a in agents if a.agent_id == agent_id), None)
                if source:
                    print(f"Agent {agent_id} broadcasts (scheduled): {announcement}")
                    # Broadcast to direct neighbors.
                    if agent_id in connections:
                        for nb_id in connections[agent_id]:
                            neighbor = next((a for a in agents if a.agent_id == nb_id), None)
                            if neighbor:
                                neighbor.update_epistemic_state(announcement)
                                print(f"Agent {nb_id} receives scheduled broadcast from Agent {agent_id}: {announcement}")
        
        # Process any predefined epistemic action for this tick.
        if tick in actions:
            agent_id, announcement = actions[tick]
            source = next((a for a in agents if a.agent_id == agent_id), None)
            if source:
                source.update_epistemic_state(announcement)
                print(f"Agent {agent_id} performs predefined action: {announcement}")
                # Immediately broadcast the predefined action to direct neighbors.
                if agent_id in connections:
                    for nb_id in connections[agent_id]:
                        neighbor = next((a for a in agents if a.agent_id == nb_id), None)
                        if neighbor:
                            neighbor.update_epistemic_state(announcement)
                            print(f"Agent {nb_id} receives predefined broadcast from Agent {agent_id}: {announcement}")
        else:
            print("No predefined epistemic action this tick.")
        
        # For each agent, check if they have "learned" a proposition.
        # If for a proposition p (in {"contract", "a_contract", "b_contract"})
        # every world in their epistemic state satisfies p = True, and they haven't yet broadcast it,
        # then schedule a broadcast for the next tick.
        for agent in agents:
            for prop in ["contract", "a_contract", "b_contract"]:
                if prop not in agent.broadcasted:
                    if all(agent.domain[w][prop] for w in agent.possible_worlds):
                        next_tick = tick + 1
                        if next_tick not in scheduled_broadcasts:
                            scheduled_broadcasts[next_tick] = []
                        scheduled_broadcasts[next_tick].append((agent.agent_id, {prop: True}))
                        agent.broadcasted.add(prop)
                        print(f"Agent {agent.agent_id} schedules broadcast of {{'{prop}': True}} for tick {next_tick}.")
        
        # Visualize the combined Kripke model after all updates in this tick.
        visualize_combined_kripke(agents, domain, tick)
        print("\n=== CURRENT EPISTEMIC STATES ===")
        for a in agents:
            print(str(a))
    
    # Final summary.
    print("\n=== FINAL EPISTEMIC STATES ===")
    for a in agents:
        print(str(a))

if __name__ == "__main__":
    epistemic_simulation(num_agents=10, num_ticks=5, p_connection=0.4)
```
